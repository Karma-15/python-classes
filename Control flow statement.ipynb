Conditionals(if,elif,else)
Conditionals let a programe make decisions.
"If something is true -> do this. otherwise -> do that"

Basic syntax:
if condition:
    runs if condition is true
elif another_condition:
    runs if above is false and this is true
else:
    runs if all conditions are false

```python
# simple if statement

age=20
if age>=18:
    print("you can vote")
```

```python
# if-else statement

rain=False

if rain:
    print("Take umbrella")
else:
    print("don't take umbrella")
```

```python
# elif statement

weather="sunny"

if weather=="rain":
    print("Take umbrella")
elif weather=="sunny":
    print("Take sunglasses")
else:
    print("It is a pleasent day")
```

```python
# if else

marks=35

if marks>=40:
    print("Pass")
else:
    print("Fail")
```

```python
#  if elif else

score=78

if score>=90:
    print("Grade A")
elif score>=75:
    print("Grade B")
elif score>=60:
    print("Grade C")
else:
    print("Grade D")
```

```python
# checking temperature

# use if else and print if it is hot if temperature is >25 print not hot

temperature=20
if temperature>=40:
    print("Hot")
else:
    print("Not Hot")
```

```python
# even or odd 

num=8
if num%2==0:
    print("It is even")
else:
    print("It is odd")
```

```python
# login check

username="admin"
password="1234"

if username=="admin" and password=="1234":
    print("Login Successful")
else:
    print("Login Failed")
```

structural pattern matching (match-case)

March-case is python's switch statement

use it when:
you compare one variable against many fixed values

Basic Syntax:

match value:
    case option1:
        #code
    case option2:
        #code
    case option3:
        #code
    case _:
        #default

```python
# menu selection

choice=int(input("Enter your choice"))

match choice:
    case 1:
        print("Add item")
    case 2:
        print("Delete item")
    case 3:
        print("Update item")
    case _:
        print("Invalid choice")
```

```python
# using strings

day="Sunday"

match day:
    case "Saturday"|"Sunday":
        print("Weekend")
    case _:
        print("Weekday")
```

```python
# simple calcultor option

operation="+"

match operation:
    case"+":
        print("Addition selected")                           # match -> check this value
    case"-":                                                 # case  -> if it matches this
        print("Subtraction selected")                        # _     -> anything else
    case"*":
        print("Multiplication selected")
    case"/":
        print("Division selected")
    case _:
        print("Invalid operation")
    
```

match is not for condition like greater than or less than
match cannot be used for a range of values either

case age>18 #invalid
case num>30 #invalid
