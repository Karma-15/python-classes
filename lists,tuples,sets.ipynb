```python
# lists methods - (.append, .insert, .remove, .pop)

list2=[10,20,30,40]
list2.append(50)          # append refers to adding the value at the end of the list
print(list2)

list2=[10,20,30,40]
list2.append([50,60,70])          
print(list2)

list2.insert(1,36)        # insert refers to add a value using index position 
print(list2)

list2.remove(36)          # remove refers to the extract value removed from the list
print(list2)

list2.pop(1)              # pop refers to the removing element using index position
print(list2)

list2.pop()              # pop refers to the removing the last value
print(list2)
```

```python
# set methods 

# adding elements

colors=set(["red","green","blue"])
colors.add("yellow")
print(colors)
colors.update(["black","white"])
print(colors)
```

```python
# removing elements

print(colors)
colors.remove("blue")
print(colors)
```

```python
# removing random elements

print(colors)
colors.pop()
print(colors)
colors.clear()     # removes all the elements
print(colors)
```

```python
a={1,2,3}
b={4,5,6}

# union
# a | b

print(a.union(b))                        # it gives all union elements

# intersection
# a & b

print(a.intersection(b))                 # common elements

# difference
# a - b

print(a.difference(b))                   # elements in A not in a

# symmetric difference
# a ^ b

print(a.symmetric_difference(b))         # elements in either but not both

```

```python
# Dictionaries- 

student={"name":"Alex","age":22,"marks":85}
print(student["name"])                      # calling and accessing the data through key instead of index
```

```python
# Dictionary methods 

# Adding / Updating a dictionary

student["city"]="Banglore"
student["Age"]=23
print(student)

# removing items

student.pop("marks")     # remove key
student.popitem()        # remove last item
del student["age"]
print(student)
```

```python
# Keys

print(student.keys())      # Gives only keys 
print(student.values())    # Gives only values
print(student.items())     # Entire dictionary
```

BITWISE OPERATORS

Binary and &     - both bits if 1=1
Binary or |      -either bits must be 1
Binary xor ^     -either one of the bits only must be one
negation ~       -if the bit is one then output is 0 and if bit is 0 then output is 1
left shift <<    -shifting towards left and adding zeros at the right
right shift >>   -shifting towards right and dropping bits

Bitwise   &    --  T & T  --> T
                   T & F  --> F
                   F & T  --> F
                   F & F  --> F

Bitwise   |    --  T | T --> T
                   T | F --> T 
                   F | T --> T
                   F | F --> F

Bitwise   ^    --  T ^ T --> F
                   T ^ F --> T 
                   F ^ T --> T
                   F ^ F --> F

Negation ~  --> ~x=-(x+1)  2's complement formula



```python
print(5&3)
print(5|3)
print(5^3)
print(~5)
print(5<<1)
print(5>>1)
```
