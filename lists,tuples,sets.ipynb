Method	                        Meaning	                      Example	                 Output
union()	                 Combines elements from both	        A.union(B)	         {1,2,3,4,5,6,7,8}
intersection()	             Common elements	                A.intersection(B)	      {4,5}
difference()	         Elements only in A	                 A.difference(B)	         {1,2,3}
symmetric_difference()	   Elements not in both	          A.symmetric_difference(B)	   {1,2,3,6,7,8}
update()	            Adds items from another set to A	    A.update(B)	            modifies A
difference_update()	      Removes common elements	         A.difference_update(B)	    modifies A
issubset()	           True if all of A in B	           {1,2}.issubset(A)	           True
issuperset()	       True if A contains all of B	         A.issuperset({1,2})	       True
isdisjoint()	       True if no common elements	        {9,10}.isdisjoint(A)	        True

```python
# set operations

a={1,2,3}
b={4,5,6}

# union
# a | b

print(a.union(b))                        # it gives all union elements

# intersection
# a & b

print(a.intersection(b))                 # common elements

# difference
# a - b

print(a.difference(b))                   # elements in A not in B

# symmetric difference
# a ^ b

print(a.symmetric_difference(b))         # elements in either but not both

print(a.difference_update(b))            # Removes common elements

A={1,2}
B={2,3,4,1,5}

print(A.issubset(B))                    # True if all of A in B

print(B.issuperset(A))                  # True if B contains all of A

print(A.isdisjoint(b))                  # True if no common elements

```

BITWISE OPERATORS

Binary and &     - both bits if 1=1
Binary or |      -either bits must be 1
Binary xor ^     -either one of the bits only must be one
negation ~       -if the bit is one then output is 0 and if bit is 0 then output is 1
left shift <<    -shifting towards left and adding zeros at the right
right shift >>   -shifting towards right and dropping bits

Bitwise   &    --  T & T  --> T
                   T & F  --> F
                   F & T  --> F
                   F & F  --> F

Bitwise   |    --  T | T --> T
                   T | F --> T 
                   F | T --> T
                   F | F --> F

Bitwise   ^    --  T ^ T --> F
                   T ^ F --> T 
                   F ^ T --> T
                   F ^ F --> F

Negation ~  --> ~x=-(x+1)  2's complement formula



```python
print(5&3)
print(5|3)
print(5^3)
print(~5)
print(5<<1)
print(5>>1)
```
